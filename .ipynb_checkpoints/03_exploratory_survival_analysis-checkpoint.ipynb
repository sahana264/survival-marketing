## 1. Kaplan–Meier Estimator

### 1.1 Definition

At event times $(t_{(1)} < t_{(2)} < \dots)$:

$$
\hat S(t) = \prod_{t_i \le t}\!\left(1 - \frac{d_i}{n_i}\right)
$$

where  
- $n_i$: number at risk just before $t_i$  
- $d_i$: number of events at $t_i$

Each term $(1 - d_i/n_i)$ is the **conditional probability of surviving past** that event time.

The product of all such terms up to time $t$ gives the **overall survival probability** up to $t$.

---

### 1.1 Derivation (heuristic) and Intuition

Survival can be written as the product of conditional survival probabilities (chain rule of probability):

$$
P(T > t_k) = \prod_{i=1}^{k} P(T > t_i \mid T \ge t_i)
$$

Each conditional survival probability is estimated empirically by

$$
P(T > t_i \mid T \ge t_i) \approx 1 - \frac{d_i}{n_i}.
$$

Multiplying over all observed event times gives

$$
\hat S(t) = \prod_{t_i \le t}\!\left(1 - \frac{d_i}{n_i}\right).
$$

This is the **Kaplan–Meier (product-limit) estimator**, the **non-parametric maximum-likelihood estimator** of the survival function $S(t)$ under **independent right-censoring**. 

> Kaplan–Meier estimator can be derived from maximum likelihood estimation of the discrete hazard function.
> https://en.wikipedia.org/wiki/Kaplan%E2%80%93Meier_estimator

---

### 1.2 Historical Note

Published in 1958 by **Edward L. Kaplan** and **Paul Meier**, the Kaplan–Meier estimator is regarded as one of the **most important statistical innovations of the 20ᵗʰ century** with over 35,000 citations.

> The estimator is named after Edward L. Kaplan and Paul Meier, who each submitted similar manuscripts to the Journal of the American Statistical Association. The journal editor, John Tukey, convinced them to combine their work into one paper, which has been cited more than 34,000 times since its publication in 1958
> -https://en.wikipedia.org/wiki/Kaplan%E2%80%93Meier_estimator

Their product-limit method made it possible to estimate survival probabilities **without assuming any distribution**, even when many observations are censored.

Its core idea - **extracting structure from incomplete data** - underpins much of modern survival analysis including time-to-event modeling in marketing.

### 1.3 Kaplan–Meier Computation Example by Hand

We now compute the Kaplan–Meier survival function step-by-step using the small dataset from Section 0.

At each **event time**, we calculate:

- $n_i$: number of subjects still at risk just before time $t_i$
- $d_i$: number of events (failures/churns) at that time  
- $(1 - d_i/n_i)$: conditional probability of surviving past that time  
- $\hat S(t_i)$: cumulative survival probability up to that point, given by multiplying all previous conditional survivals  

---

| Time $(t_i)$ | At risk $(n_i)$ | Events $(d_i)$ | Calculation | Survival $(\hat S(t_i))$ | Notes |
|:--:|:--:|:--:|:--|:--:|:--|
| 1 | 8 | 1 | $(1 - \frac{1}{8} = 0.875)$ | $(1.0 \times 0.875 = 0.875)$ | One event out of 8 at risk |
| 2 | 7 | 2 | $(1 - \frac{2}{7} = 0.714)$ | $(0.875 \times 0.714 = 0.625)$ | Two events (tie) |
| 3 | 5 | 0 | – | 0.625 | Only censoring → survival unchanged |
| 4 | 4 | 0 | – | 0.625 | Only censoring → survival unchanged |
| 5 | 3 | 1 | $(1 - \frac{1}{3} = 0.667)$ | $(0.625 \times 0.667 = 0.417)$ | One event → sharp drop |
| 7 | 2 | 1 | $(1 - \frac{1}{2} = 0.5)$ | $(0.417 \times 0.5 = 0.208)$ | One event → large drop |
| 8 | 1 | 1 | $(1 - 1 = 0)$ | $(0.208 \times 0 = 0)$ | Last subject fails → survival reaches 0 |

### 2.1  Visualizing the KM Curve

We visualize the KM curve for the data in the table above using a simple Matplotlib step plot.

```python
import numpy as np
import matplotlib.pyplot as plt

# Event times and KM survival probabilities
times = np.array([0, 1, 2, 3, 4, 5, 7, 8])
S = np.array([1.0, 0.875, 0.625, 0.625, 0.625, 0.417, 0.208, 0.0])

# Step plot
plt.figure(figsize=(7,4))
plt.step(times, S, where='post', linewidth=2)
plt.scatter(times, S, color='black', zorder=3)

# Labels
plt.title("Kaplan–Meier Survival Curve (Manual Example)")
plt.xlabel("Time")
plt.ylabel("Survival probability $\\hat{S}(t)$")
plt.grid(alpha=0.3)
plt.show()
```

Let us now write a function to plot the survival curve. The code computes and plots the K-M survival curves. It works by:
1. Sorting the data by time.
2. At each time-point, calculates how many individuals were at risk $n_i$ and how many experienced the event $d_i$.
3. Updates the survival probability by multiplying the previous survival probability with $(1-\frac{d_i}{n_i})$.
4. Repeats this for all event times to build the stepwise survival curve.

```python
import pandas as pd

def km_plot(df, duration_col, event_col, title="K-M Survival Curve"):
    """
    Computes and plots K-M survival curve.

    df: pandas.DataFrame
    Dataset containing duration and event columns.

    duration_col: str
    Column name for time-to-event durations.

    event_col: str
    Column name (1 if event occurred, 0 if censored).

    title : str
    Plot title.
    """

    # sort by time
    df = df.sort_values(duration_col)
    # get unique event times
    event_times = np.sort(df.loc[df[event_col] == 1, duration_col].unique())

    #initialize
    surv = 1.0
    surv_probs = [1.0]
    time_points = [0.0]

    for t in event_times:
        # number at risk just before time t (e.g.,customers who are active at time t or beyond)
        n_i = (df[duration_col]>=t).sum() 
        # number of events at t
        d_i = ((df[duration_col]==t) & (df[event_col]==1)).sum()
        surv *= (1-d_i/n_i)
        surv_probs.append(surv)
        time_points.append(t)

    # convert to dataframe
    km_df = pd.DataFrame({"time":time_points, "survival":surv_probs})

    #plot
    plt.figure(figsize=(8, 6))
    plt.step(km_df["time"], km_df["survival"], where="post", label="KM-Estimate")
    plt.title(title, fontsize=12)
    plt.xlabel("Time", fontsize=12)
    plt.ylabel("Survival Probability", fontsize=12)
    plt.grid(alpha=0.3)
    plt.legend()
    plt.show()

# Let us put our data in a dataframe
df_example = pd.DataFrame({
    "id": [1, 2, 3, 4, 5, 6, 7, 8],
    "time": [1, 2, 4, 7, 2, 3, 5, 8],
    "event": [1, 1, 0, 1, 1, 0, 1, 1]  # 1 = event occurred, 0 = censored
})

km_plot(df_example, duration_col="time", event_col="event")
```

## 3. Greenwood’s Variance

The Kaplan–Meier estimator gives the estimated survival probability $\hat S(t)$  
To measure uncertainty, we use **Greenwood’s formula** to estimate its variance.

Think of it like this : 
- the KM curve shows your best guess for retention over time.
- The confidence interval shows your margin of error, how sure you are about that retention level.

<div style="background-color:#f0f7ff; border-left:4px solid #1890ff; padding:10px; border-radius:8px;">
    
### 3.1 Formula and Intuition (Optional)

**Step 1. Write the KM estimator and take logs**

$$
\hat S(t)=\prod_{t_j\le t}\Big(1-\frac{d_j}{n_j}\Big)
\quad\Rightarrow\quad
\log\hat S(t)=\sum_{t_j\le t}\log\!\Big(1-\frac{d_j}{n_j}\Big)
$$

**Reason:**  
The product becomes a sum, which makes the variance easier to handle.

**Step 2. Model each event count**

At time $(t_j)$:  
$$
d_j \sim \mathrm{Binomial}(n_j, p_j)
$$

So define $(X_j = \frac{d_j}{n_j})$, giving  

$$
E(X_j)=p_j,\quad \operatorname{Var}(X_j)=\frac{p_j(1-p_j)}{n_j}.
$$

**Reason:**  
Each of the $(n_j)$ individuals at risk has probability $(p_j)$ of failing at $(t_j)$.

**Step 3. Linearize each log-term (Delta Method / 1st-order Taylor)**

Expand $(\log(1 - X_j))$ around $(p_j)$:

$$
\log(1 - X_j) \approx \log(1 - p_j) - \frac{1}{1 - p_j}(X_j - p_j)
$$

**Reason:**  
First-order Taylor expansion gives a linear approximation suitable for variance computation.

**Step 4. Compute variance of a single term**

$$
\operatorname{Var}\!\big[\log(1 - X_j)\big] \approx 
\left(\frac{1}{1 - p_j}\right)^2 \operatorname{Var}(X_j)
= \frac{p_j}{n_j(1 - p_j)}.
$$

**Reason:**  
Apply variance to the linear term and substitute \(\operatorname{Var}(X_j)\).

**Step 5. Sum the variances (ignore covariance terms)**

Although $(X_j)$ values are not independent,  
$$
E\!\left[\frac{d_j}{n_j} - p_j \mid d_1, \ldots, d_{j-1}\right] = 0,
$$
which implies the covariance between different terms is 0.  
Thus,

$$
\operatorname{Var}\!\Big[\sum_{t_j \le t} \log(1 - X_j)\Big] 
= \sum_{t_j \le t} \operatorname{Var}\!\big[\log(1 - X_j)\big].
$$

**Reason:**  
Conditional zero mean ⇒ covariances vanish ⇒ variances add up.

**Step 6. Substitute the expression for variance**

$$
\operatorname{Var}[\log \hat S(t)] \approx 
\sum_{t_j \le t} \frac{p_j}{n_j(1 - p_j)}.
$$

Replace $(p_j)$ by the estimate $(\hat p_j = \frac{d_j}{n_j})$:

$$
\operatorname{Var}[\log \hat S(t)] \approx 
\sum_{t_j \le t} \frac{d_j}{n_j (n_j - d_j)}.
$$

**Reason:**  
Use plug-in estimate for the true event probability.

**Step 7. Use Delta Method to get variance of $(\hat S(t))$**

Since $(\hat S(t) = \exp(\log \hat S(t)))$,  
$(f(x) = e^x \Rightarrow f'(x) = e^x)$.

$$
\operatorname{Var}[\hat S(t)] \approx 
[\hat S(t)]^2 \operatorname{Var}[\log \hat S(t)].
$$

Substitute the previous step:

$$
\boxed{
\operatorname{Var}[\hat S(t)] = [\hat S(t)]^2 
\sum_{t_j \le t} \frac{d_j}{n_j (n_j - d_j)}
}
$$

**Reason:**  
The delta method transforms variance from the log scale back to the survival scale.


### **Final Intuition**

Each event time contributes uncertainty $(\frac{d_j}{n_j(n_j - d_j)})$, and since survival probabilities multiply over time, their variances compound multiplicatively - hence the $([\hat S(t)]^2)$ term.


**References**:
> The variance of the Kaplan–Meier survival estimate is computed using Greenwood’s formula (Greenwood, 1926), as described in Craig (2020) — Lecture Notes on Survival Analysis, Purdue University.
> - https://www.stat.purdue.edu/~bacraig/notes526/topic15a.pdf

> In statistics, the delta method is a method of deriving the asymptotic distribution of a random variable. It is applicable when the random variable being considered can be defined as a differentiable function of a random variable which is asymptotically Gaussian.
> The intuition of the delta method is that any such g function, in a "small enough" range of the function, can be approximated via a first order Taylor series (which is basically a linear function). If the random variable is roughly normal then a linear transformation of it is also normal. Small range can be achieved when approximating the function around the mean, when the variance is "small enough". When g is applied to a random variable such as the mean, the delta method would tend to work better as the sample size increases, since it would help reduce the variance, and thus the Taylor approximation would be applied to a smaller range of the function g at the point of interest.
> -https://en.wikipedia.org/wiki/Delta_method

### 3.2 Step-by-Step Calculation (Using Our Example)

We look at the same table:

| id | time | event |
|:--:|:----:|:-----:|
| 1  | 1    | 1     |
| 2  | 2    | 1     |
| 3  | 4    | 0     |
| 4  | 7    | 1     |
| 5  | 2    | 1     |
| 6  | 3    | 0     |
| 7  | 5    | 1     |
| 8  | 8    | 1     |

#### Compute KM and Greenwood variance

| $(t_j)$ | $(n_j)$ | $(d_j)$ | factor $(1-\dfrac{d_j}{n_j})$ | $(\hat S(t_j))$ (KM, fraction) | Greenwood term $(c_j=\dfrac{d_j}{n_j(n_j-d_j)})$ |
|:-------:|:-------:|:-------:|:-----------------------------:|:-----------------------------:|:-----------------------------------------------:|
| 1       | 8       | 1       | $(1-\tfrac{1}{8}=\tfrac{7}{8})$     | $(\hat S(1)=\tfrac{7}{8})$         | $(\dfrac{1}{8\cdot7}=\dfrac{1}{56})$|
| 2       | 7       | 2       | $(1-\tfrac{2}{7}=\tfrac{5}{7})$     | $(\hat S(2)=\tfrac{7}{8}\cdot\tfrac{5}{7}=\tfrac{5}{8})$ | $(\dfrac{2}{7\cdot5}=\dfrac{2}{35})$            |
| 5       | 3       | 1       | $(1-\tfrac{1}{3}=\tfrac{2}{3})$     | $(\hat S(5)=\tfrac{5}{8}\cdot\tfrac{2}{3}=\tfrac{5}{12})$ | $(\dfrac{1}{3\cdot2}=\dfrac{1}{6})$             |
| 7       | 2       | 1       | $(1-\tfrac{1}{2}=\tfrac{1}{2})$     | $(\hat S(7)=\tfrac{5}{12}\cdot\tfrac{1}{2}=\tfrac{5}{24})$ | $(\dfrac{1}{2\cdot1}=\dfrac{1}{2})$             |


#### Greenwood variance ingredients (exact)

- Sum of Greenwood terms up to $(t=7)$:
  $$
  \sum_{j\le 7} c_j
  = \frac{1}{56} + \frac{2}{35} + \frac{1}{6} + \frac{1}{2}
  = \frac{15}{840} + \frac{48}{840} + \frac{140}{840} + \frac{420}{840}
  = \frac{623}{840}.
  $$

- So (exact):
  $$
  \operatorname{Var}[\log \hat S(7)] \approx \frac{623}{840}.
  $$

- KM at \(t=7\):
  $$
  \hat S(7)=\frac{5}{24}.
  $$

---

#### 95% CI for $(\hat S(t))$ (using normal approx on $(\log \hat S(t)))$

**Method (delta on log-scale):**  
$$
\text{SE}_{\log} = \sqrt{\sum c_j} = \sqrt{\frac{623}{840}}.
$$
Use $(z_{0.975}=1.96)$.  Then
$$
\log\hat S(7) \pm 1.96\cdot\text{SE}_{\log}
\quad\Rightarrow\quad
\text{CI for }\hat S(7) = \exp\big( \log\hat S(7) \pm 1.96\sqrt{623/840} \big).
$$

**Plugging values (numerical evaluation shown for the CI step):**
- $(\hat S(7)=\dfrac{5}{24})$.  
- $(\log\hat S(7)=\log(5)-\log(24))$  
- $(\dfrac{623}{840}\approx 0.742857)$, $(\sqrt{623/840}\approx 0.861023)$ 
- margin $(=1.96\times 0.861023\approx 1.6866)$

So
$$
\log\hat S(7)\approx -1.568616 \quad\Rightarrow\quad
\begin{cases}
\text{lower log} = -1.568616 - 1.6866 \approx -3.2552,\\[4pt]
\text{upper log} = -1.568616 + 1.6866 \approx 0.1180.
\end{cases}
$$

Exponentiate to get the CI for $(\hat S(7))$:
$$
\begin{aligned}
\text{lower} &= e^{-3.2552}\approx 0.0385,\\
\text{upper} &= e^{0.1180}\approx 1.1252\ (\text{often truncated to }1).
\end{aligned}
$$

#### Final 
- KM at $(t=7)$: $(\displaystyle \hat S(7)=\frac{5}{24})$  
- Greenwood sum: $(\displaystyle \sum c_j=\frac{623}{840})$  
- 95% CI (log-scale delta method):  
  $$
  \boxed{\; \hat S(7)=\tfrac{5}{24},\quad 
  \text{95\% CI}\approx \big(0.0385,\ \min(1,\,1.1252)\big)= (0.0385,\ 1)\;}
  $$



### 3.4 Interpretation

- Greenwood’s variance quantifies **uncertainty in the survival estimate**.  
- It grows over time as fewer subjects remain at risk.  
- In survival plots, the shaded **confidence band** around the KM curve comes from this formula.  
- When comparing groups, overlapping confidence bands often indicate **no significant difference** in survival.

### 3.5 KM plot with confidence intervals

Now we plot the K-M using the Greenwood's variance (log-log CI).

```python
# confidence band around K-M curve

from math import sqrt
from scipy.stats import norm

def km_plot_greenwood(df, duration_col, event_col, alpha=0.05, title="Kaplan–Meier with Greenwood CI"):
    """
    Compute and plot the Kaplan–Meier survival curve with Greenwood standard error and 95% CI.

    Parameters
    ----------
    df : pandas.DataFrame
        Must contain time and event columns.
    time_col : str
        Column name with time-to-event durations.
    event_col : str
        Column name with event indicator (1 = event occurred, 0 = censored).
    alpha : float
        Significance level (default 0.05 for 95% CI).
    title : str
        Plot title.
    """

    # sort data by time
    df = df.sort_values(duration_col)

    # unique event times
    event_times = np.sort(df.loc[df[event_col] == 1, duration_col].unique())

    # initialize
    surv = 1.0
    var = 0.0
    surv_list = [1.0]
    time_list = [0.0]
    lower_ci = [1.0]
    upper_ci = [1.0]
    z = norm.ppf(1 - alpha/2)

    for t in event_times:
        # number at risk just before time t
        n_i = (df[duration_col] >= t).sum()
        # number of events at time t
        d_i = ((df[duration_col] == t) & (df[event_col] == 1)).sum()

        # Update survival probability
        surv *= (1 - d_i / n_i)

        # Greenwood’s variance accumulation
        var += d_i / (n_i * (n_i - d_i)) if n_i > d_i else 0
        se = surv * sqrt(var)

        # compute log-log CI (standard in KM)
        if surv > 0:
            loglog = np.log(-np.log(surv))
            se_loglog = sqrt(var) / abs(np.log(surv))
            ci_lower = np.exp(-np.exp(loglog + z * se_loglog))
            ci_upper = np.exp(-np.exp(loglog - z * se_loglog))
        else:
            ci_lower, ci_upper = np.nan, np.nan

        time_list.append(t)
        surv_list.append(surv)
        lower_ci.append(ci_lower)
        upper_ci.append(ci_upper)

    # Build results DataFrame
    km_df = pd.DataFrame({
        "time": time_list,
        "survival": surv_list,
        "lower_ci": lower_ci,
        "upper_ci": upper_ci
    })

    # --- Plotting ---
    plt.figure(figsize=(8, 6))
    plt.step(km_df["time"], km_df["survival"], where="post", label="KM Estimate", linewidth=2)
    plt.fill_between(km_df["time"], km_df["lower_ci"], km_df["upper_ci"],
                     color="skyblue", alpha=0.3, step="post", label="95% CI")

    plt.title(title, fontsize=14)
    plt.xlabel("Time", fontsize=12)
    plt.ylabel("Survival Probability", fontsize=12)
    plt.grid(alpha=0.3)
    plt.legend()
    plt.show()

    return km_df

# applying in to our data

km_plot_greenwood(df_example, duration_col="time", event_col="event")
```
